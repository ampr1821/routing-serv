```python
def roundoff(num):
    x = format(num, '.8f')
    if int(x[-1]) < 5:
        done = int(x[-2])
    else:
        done = int(x[-2])+1
    return float(format(num,'.6f')+str(done))
```

```python
import json

def extract_data(json_data):
    results = json_data.get('results', [])
    extracted_data = []
    for result in results:
        links = result.get('location', {}).get('shape', {}).get('links', [])
        current_flow = result.get('currentFlow', {})
        speed = current_flow.get('speed')
        free_flow = current_flow.get('freeFlow')
        jam_factor = current_flow.get('jamFactor')
        ratio_of_speed_to_free_flow = speed / free_flow if free_flow != 0 else 0
        for link in links:
            points = link.get('points', [])
            for i in range(len(points)-1):
                lat1 = points[i]['lat']
                lng1 = points[i]['lng']
                lat2 = points[i+1]['lat']
                lng2 = points[i+1]['lng']
                lat1 = roundoff(float(lat1))
                lat2 = roundoff(float(lat2))
                lng1 = roundoff(float(lng1))
                lng2 = roundoff(float(lng2))
                extracted_data.append((lat1, lng1, lat2, lng2, jam_factor, ratio_of_speed_to_free_flow))
    return extracted_data

# Open the file and read the contents into a string
with open('response2.json') as f:
    json_text = f.read()

# Parse the JSON string into a Python object
json_data = json.loads(json_text)

# Call the extract_data function with the parsed JSON object
result = extract_data(json_data)

# Write the extracted data to a CSV file
with open('output.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['lat1', 'lng1', 'lat2', 'lng2', 'jam_factor', 'ratio_of_speed_to_free_flow'])
    for row in result:
        writer.writerow(row)

# Print a message indicating that the file was written successfully
print('Data written to output.csv')
```
